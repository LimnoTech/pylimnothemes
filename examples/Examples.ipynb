# pylimnothemes usage guide 

This notebook demonstrates how to use the pylimnothemes library. Please see the readme for details on installing the library.

# Install Packages

```python
import pylimnothemes
from pylimnothemes import *

import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
```

# Create A Simple Figure

## Matplotlib Defaults

```python
# evenly sampled x values
t = np.arange(0., 5., 0.2)

# plot it
plt.plot(t, t, label='linear')
plt.plot(t, t**2, label = 'quadratic')
plt.plot(t, t**3, label = 'cubic')
plt.plot(t, t**4, label = 'biquadrates')
plt.ylabel("Values")
plt.xlabel("Time")
plt.title("Values over Time")
plt.legend()
plt.show()
```

## Pylimnotheme: Main Palette

Call `theme_limno`. The main color palette is the default, so no input arguments are required. 

```python
theme_limno()
```

```python
mpl.rcParams['text.color'] = '#999999'
mpl. rcParams['axes.edgecolor']  = '#999999'
plt.plot(t, t, label='linear')
plt.plot(t, t**2, label = 'quadratic')
plt.plot(t, t**3, label = 'cubic')
plt.plot(t, t**4, label = 'biquadrates')
plt.ylabel("Values")
plt.xlabel("Time")
plt.title("Values over Time")
plt.legend()
plt.show()
```

```python
theme_limno(reverse=True)
```

## Pylimnotheme: Secondary Palette

Call `theme_limno` again, but now must specify that `palette = 'secondary'`

```python
theme_limno(palette='secondary')
```
